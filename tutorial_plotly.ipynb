## Package Structure of Plotly

Existem três módulos principais em Plotly. Eles são:

    plotly.plotly
    plotly.graph.objects
    plotly ferramentas.

* __plotly.plotly__ atua como a interface entre a máquina local e Plotly. Ele contém funções que requerem uma resposta do servidor do Plotly.

* __plotly.graph_objects__ módulo contém os objetos (Figura, layout, dados e a definição dos gráficos como gráfico de dispersão, gráfico de linhas) que são responsáveis pela criação dos gráficos.  A Figura pode ser representada como dict ou instâncias de __plotly.graph_objects__.Figure e estes são serializados como JSON antes de ser passado para plotly.js. Considere o exemplo abaixo para uma melhor compreensão.

Nota: o módulo plotly.express pode criar a Figura inteira de uma só vez. Usa os graph_objects internamente e devolve a instância graph_objects.figure.

Exemplo:

```python
import plotly.express as px

# Criando a figura 
fig = px.line(x=[1, 2, 3], y =[1, 2, 3])

# plotando gráfico
print(fig)
```

As figuras são representadas como árvores onde o nó raiz tem três atributos de camada superior - dados, disposição e quadros e os nós nomeados chamados 'atributos'. Considere o exemplo acima, layout.legend é um dicionário aninhado onde a legenda é a chave dentro do dicionário cujo valor também é um dicionário.  

plotly.tools módulo contém várias ferramentas nas formas das funções que podem melhorar a experiência Plotly.

## Criando diferentes tipos de gráficos

Com plotly podemos criar mais de 40 gráficos e cada gráfico pode ser criado usando a classe plotly.express e plotly.graph. Vamos ver alguns gráficos comumente usados com a ajuda de Plotly.

### Line Chart
O __gráfico de linha__(Line plot) em Plotly é muito acessível e ilustre anexação a plotly que gerencia uma variedade de tipos de dados e monta estatísticas fáceis de estilo. Com `px.line` cada posição de dados é representada como um vértice  (que localização é dada pelas colunas x e y) de uma marca de polilinha no espaço 2D.

Exemplo:

```python
# exemplo 02
import plotly.express as px

# usando dataset Iris
df = px.data.iris()

# plotting the line chart
fig = px.line(df, x="species", y="petal_width")

# showing the plot 
fig.show()
```

### Gráfico de barras (Bar Plot)

Um gráfico de barras é uma representação pictórica de dados que apresenta dados categóricos com barras retangulares com alturas ou comprimentos proporcionais aos valores que eles representam. Em outras palavras, é a representação pictórica do conjunto de dados. Esses conjuntos de dados contêm os valores numéricos das variáveis que representam o comprimento ou a altura.

Exemplo:


```python
import plotly.express  as px

# using the iris dataset
df = px.data.iris()


# plotando gráfico 
fig = px.bar(df, x='sepal_width', y="sepal_length") # largura da sépala e comprimento sépala

# mostrando o plot 
fig.show()
```

### Histogramas

Um histograma contém uma área retangular para exibir a informação estatística que é proporcional à frequência de uma variável e sua largura em intervalos numéricos sucessivos. Uma representação gráfica que gerencia um grupo de pontos de dados em diferentes intervalos especificados. Tem uma característica especial que não mostra lacunas entre as barras e semelhante a um gráfico de barras vertical.

Exemplo:


```python
import plotly.express as px

# usando dataser Iris 
df = px.data.iris()

# plotando gráfico 
fig = px.histogram(df, x="sepal_length", y="petal_width")

# mostrando o plot 
fig.show()
```

Gráfico de dispersão e gráficos de bolhas

* Um __scatter plot__ é um conjunto de pontos pontilhados para representar partes individuais de dados nos eixos horizontal e vertical. Um gráfico no qual os valores de duas variáveis são plotados ao longo do eixo X e do eixo Y, o padrão dos pontos resultantes revela uma correlação entre eles.

* A __bubble plot__ é um gráfico de dispersão com bolhas (círculos coloridos). As bolhas têm vários tamanhos dependentes de outra variável nos dados. Ele pode ser criado usando o método scatter() de plotly.express.

Exemplo 1: Gráfico de dispersão


```python
import plotly.express as px

# usando dataset Iris 
df = px.data.iris()

# plotando gráfico 
fig = px.scatter(df, x="species", y="petal_width")

# mostrando plot 
fig.show()
```

Example 2: Bubble Plot

```python
import plotly.express as px

# usando dataset iris 
df = px.data.iris()

# plotting the bubble chart 
fig = px.scatter(df, x="species", y="petal_width",
                size="petal_length", color="species")

# mostrando plot
fig.show()
```

### Gráficos circulares

Um gráfico de pizza é um gráfico estatístico circular, que é dividido em fatias para ilustrar proporções numéricas. Ele representa um gráfico especial que usa "fatias de pizza", onde cada setor mostra os tamanhos relativos dos dados. Um gráfico circular corta em forma de raios em segmentos descrevendo frequências relativas ou magnitude também conhecida como grafo circular.

Exemplo:


```python
import plotly.express as px 

# usando dataser tips 
df = px.data.tips() # Index(['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']

# plotando Gráfica
fig = px.pie(df, values="total_bill", names='day')


# mostrando plot 
fig.show()
```

### Box Plots

Um gráfico de caixa também é conhecido como gráfico de Whisker é criado para exibir o resumo do conjunto de valores de dados com propriedades como mínimo, primeiro quartil, mediana, terceiro quartil e máximo. No gráfico de caixa, uma caixa é criada do primeiro quartil para o terceiro quartil, uma linha vertical também está lá que atravessa a caixa na mediana. Aqui o eixo x denota os dados a serem plotados enquanto o eixo y mostra a distribuição de frequência.

Exemplo:

```python
import plotly.express as px 

# usando o dataset tips 
df = px.data.tips()

# plotando gráfico
fig=px.box(df, x="day", y="total_bill", color="day")

# mostrando Gráfico 
fig.show()
```

### Gráfico tipo violino

Violin Plot é um método para visualizar a distribuição de dados numéricos de diferentes variáveis. É semelhante ao Box Plot, mas com um gráfico girado em cada lado, dando mais informações sobre a estimativa de densidade no eixo y. A densidade é espelhada e invertida e a forma resultante é preenchida, criando uma imagem semelhante a um violino. A vantagem de um enredo de violino é que ele pode mostrar nuances na distribuição que não são perceptíveis em um boxplot. Por outro lado, o boxplot mostra mais claramente os outliers nos dados.

Exemplo:


```python
import plotly.express as px 

# usanso dataset tips 
df = px.data.tips()

# plotando o gráfico tipo violino 
fig = px.violin(
    df,
    x = "day",
    y = "total_bill",
    color="day"
)

# mostrando gráfico 
fig.show()
```

### Gantt Charts

Tabela de tempo de normalização de atividade generalizada (GANTT) é o tipo de gráfico em que séries de linhas horizontais estão presentes que mostram a quantidade de trabalho feito ou produção concluída em determinado período de tempo em relação à quantidade planejada para esses projetos. 

Exemplo:


```python
import plotly.figure_factory as ff

df = [dict(Task="Job A", Start='2009-01-01', Finish='2009-02-28', Complete=10),
      dict(Task="Job B", Start='2008-12-05', Finish='2009-04-15', Complete=60),
      dict(Task="Job C", Start='2009-02-20', Finish='2009-05-30', Complete=95)]

fig = ff.create_gantt(df, colors='Viridis', index_col='Complete', show_colorbar=True)
fig.show()
```

### Heatmaps

Heatmap é definido como uma representação gráfica de dados usando cores para visualizar o valor da matriz. Neste, para representar valores mais comuns ou atividades mais altas cores mais brilhantes, basicamente cores avermelhadas são usadas e para representar valores menos comuns ou de atividade, cores mais escuras são preferidas. Heatmap também é definido pelo nome da matriz de sombreamento. 

Exemplo: 


```python
import plotly.express as px


# usando dataser iris 
df = px.data.iris()
iris_corr = df.drop('species', axis=1).corr()


#Plotando Gráfico
fig = px.imshow(iris_corr ,text_auto=True)

# mostando gráfico
fig.show()
```

### Barras de erro (Error Bars)

Para funções que representam pontos de dados 2D, como px.scatter, px.line, px.bar, etc., barras de erro são dadas como um nome de coluna que é o valor do error_x (para o erro na posição x) e error_y (para o erro na posição y). Barras de erro são a alternância de apresentação gráfica de dados e usadas em gráficos para implicar o erro ou incerteza em uma capacidade relatada.

Exemplo:


```python
import plotly.express as px 

# usando dataset iris 
df = px.data.iris()

# calculando o campo erro 
df['error'] = df['petal_length']/100

# Plotando o gráfico
fig = px.scatter(
    data_frame=df,
    x = "species",
    y = "petal_width",
    error_x="error",
    error_y="error"
)

# mostrando o gráfico
fig.show()
```

### 3D Line Plots

O gráfico de linhas em plotagem é muito acessível e a anexação ilustre em plotagem, que gerencia uma variedade de tipos de dados e monta estatísticas fáceis de estilizar. Com px.line_3d cada posição de dados é representada como um vértice  (cuja localização é dada pelas colunas x, y e z) de uma marca de polilinha no espaço 3D.

Exemplo:


```python
import plotly.express as px

# usando dataser tips 
df = px.data.tips() 

# plotando Gráfico 
fig = px.line_3d(
    data_frame=df, 
    x='sex',
    y='day', 
    z='time', 
    color='sex'
)

# mostrando Gráfico 
fig.show()
```

### scatter plot  3D Plotly

O scatter plot  3D pode plotar gráficos bidimensionais que podem ser aprimorados mapeando até três variáveis adicionais usando a semântica de parâmetros de matiz, tamanho e estilo. Todos os parâmetros controlam a semântica visual que são usados para identificar os diferentes subconjuntos. A utilização de semântica redundante poderá ser útil para tornar os gráficos mais acessíveis. Poderá ser criada com a função scatter_3d da classe plotly.exprese.

Exemplo:




```python
import plotly.express as px 

# usando dataset iris 
df = px.data.iris() 

# plotando gráfico 
fig = px.scatter_3d(
    data_frame=df,
    x='sepal_width', 
    y='sepal_length', 
    z='petal_width',
    color='species'
)

# mostrando plot 
fig.show()

```


###Interagindo com os Gráficos

Plotly fornece várias ferramentas para interagir com os gráficos, tais como a adição de menus suspensos, botões, controles deslizantes, etc. Estes podem ser criados usando o atributo de menu de atualização do layout do gráfico. Vamos ver como fazer todas essas coisas em detalhes.
Criando o Menu Suspenso no Plotly

Um menu suspenso é uma parte do botão de menu que é exibido em uma tela o tempo todo. Cada botão de menu está associado a um widget de menu que pode exibir as opções para esse botão de menu quando clicado nele. Em plotagem, existem 4 métodos possíveis para modificar os gráficos usando o método de menu de atualização.

* restyle: modificar dados ou atributos de dados
* Relayout: modificar atributos de layout
* update: modificar dados e atributos de layout
* animate: inicia ou pausa uma animação

Exemplo:




```python
import plotly.graph_objects as go 
import numpy as np 

# Criando dataset com dados randomicos 
# travando o dados 
np.random.seed(42)

# data para se plotado 
random_x = np.random.randint(1, 101, 100)
random_y = np.random.randint(1, 101, 100)

# plotando dados 
fig = go.Figure(
    data=[
        go.Scatter(
            x=random_x,
            y=random_y,
            mode='markers'
        )
    ]
)

fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar  Chart",
                    method="restyle"
                ),
            ]),
            direction="down"
        ),
    ]
)



fig.show()
```
